```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import time
from sklearn.externals.six import StringIO  
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.preprocessing import Imputer
from sklearn.metrics import confusion_matrix
from subprocess import check_output
from sklearn import tree
from IPython.display import Image as PImage
from subprocess import check_call
from PIL import Image, ImageDraw, ImageFont
import re
import matplotlib.pyplot as plt
import graphviz
import pydotplus
```

```python
train = pd.read_csv("../input/train.csv")
```

```python
train.head()
```

```python
train['Sex'][train.Sex == 'female'] = 1
train['Sex'][train.Sex == 'male'] = 0
```

```python
train
```

```python
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
columns = ['Pclass','Sex','Age','Fare','Parch','SibSp', 'Survived']
for col in columns:
    train[col] = imp.fit_transform(train[col].values.reshape(-1,1))
```

```python
def splitDataFrameIntoSmaller(df, chunkSize): 
    listOfDf = list()
    numberChunks = len(df) // chunkSize + 1
    for i in range(numberChunks):
        listOfDf.append(df[i*chunkSize:(i+1)*chunkSize])
    return listOfDf
```

```python
X1 = train[columns]
y1 = train.Survived

X_survived = X1[y1==1]
X_died = X1[y1==0]

y_survived = X_survived['Survived']
y_died = X_died.Survived

X_array_survived = splitDataFrameIntoSmaller(X_survived, int(len(X_survived)/ 5))
X_array_died = splitDataFrameIntoSmaller(X_died, int(len(X_died)/ 5))


```

```python
X_array_survived[0]
```

```python
columns = ['Pclass','Sex','Age','Fare','Parch','SibSp']

X = pd.DataFrame(data=[])
y = pd.DataFrame(data=[])

for i in range(5):
    print(i)
    X[str(i)] = pd.Series(data = pd.concat([X_array_survived[i], X_array_died[i]]))
#     X[str(i)] = X[str(i)][columns]
    y[str(i)] = pd.Series(data = pd.concat([X_array_survived[i]['Survived'], X_array_died[i]['Survived']]))



# X0 = pd.concat([X_array_survived[0], X_array_died[0]])
# X0 = X0[columns]
# y0 = pd.concat([X_array_survived[0]['Survived'], X_array_died[0]['Survived']])

# X1 = pd.concat([X_array_survived[1], X_array_died[1]])
# X1 = X1[columns]
# y1 = pd.concat([X_array_survived[1]['Survived'], X_array_died[1]['Survived']])

# X2 = pd.concat([X_array_survived[2], X_array_died[2]])
# X2 = X2[columns]
# y2 = pd.concat([X_array_survived[2]['Survived'], X_array_died[2]['Survived']])


# X3 = pd.concat([X_array_survived[3], X_array_died[3]])
# X3 = X3[columns]
# y3 = pd.concat([X_array_survived[3]['Survived'], X_array_died[3]['Survived']])


# X4 = pd.concat([X_array_survived[4], X_array_died[4]])
# X4 = X4[columns]
# y4 = pd.concat([X_array_survived[4]['Survived'], X_array_died[4]['Survived']])
    
```

```python
X = train[columns]
y = train.Survived
```

```python
kf = KFold(n_splits=5)
kf.get_n_splits(X)

for train_index, test_index in kf.split(X):
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    clf = DecisionTreeClassifier(max_depth = 3, splitter = 'best')
    clf.fit(X_train,y_train)
    clf.score(X_test, y_test)
#     with open("tree1.dot", 'w') as f:
#         f = tree.export_graphviz(clf,
#                               out_file=f,
#                               max_depth = 5,
#                               impurity = False,
#                               feature_names = X_test.columns.values,
#                               class_names = ['No', 'Yes'],
#                               rounded = True,
#                               filled= True )
        
    #Convert .dot to .png to allow display in web notebook
    check_call(['dot','-Tpng','tree1.dot','-o','tree1.png'])

    PImage("tree1.png")

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```
